## **Number plate recognition using Haar Cascade**

```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```
#  Cascade of number plate
numberplate_cascade= cv2.CascadeClassifier('/haarcascade_russian_plate_number.xml')
```

```
# Read image
img = cv2.imread('/Car_plate_2.jpg')
```

```
# Image and cascade
plate=numberplate_cascade.detectMultiScale(img,scaleFactor=1.2,minNeighbors=5,minSize=(25,25))

# Draw rectangle around defined portion
for x,y,w,h in plate:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2) 
  
# Show processed image
cv2_imshow(img)

# Show processed image after blurring
print('-' * 70,'\nImage after blurring')
blur=cv2.blur(img,(15,15))      # The more size of kernel cause more blurring 
cv2_imshow(blur)
```
